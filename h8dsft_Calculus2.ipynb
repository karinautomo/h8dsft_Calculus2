```python
import sympy as sy
import scipy.integrate as integrate
```

1. $\int \! (3x^{2}-6x+3) \, \mathrm{d}x.$

```python
# Mencari integral tanpa batas dari fungsi

x = sy.Symbol('x')
f = 3*x**2 - 6*x + 3

f.integrate(x)
```

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.patches import Polygon
%matplotlib inline

# Mendefinisikan fungsi g
def g(x):
    return 3*x**2 - 6*x + 3

# Membuat array dari nilai x dengan rentang 0 hingga 14
x = range(0, 14)

# Mendapatkan nilai y untuk setiap nilai x
y = [g(a) for a in x]

fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(linestyle = ':')

# Plot x terhadap g(x)
plt.plot(x,y, color='firebrick')

# Membuat daerah di bawah kurva yang ingin dihitung
ix = np.linspace(1, 6)
iy = g(ix)
verts = [(1,0)] + list(zip(ix, iy)) + [(6, 1)]
poly = Polygon(verts, facecolor='pink')
ax.add_patch(poly)

plt.show()
```

```python
i, e = integrate.quad(lambda x: 3*x**2 - 6*x + 3, 1, 6)
print('%.2f' % i)
```

Luas daerah di bawah kurva $y = 3x^{2}-6x+3$ dari $x = 1$ hingga $x = 6$ adalah:

$$\int_1^6 (3x^{2}-6x+3) \, \mathrm{d}x = x^3 - 3x^2 + 3x\ \big|_1^6 = 125.00$$

2. $\int \! (8x^{3}-x^{2}+5x-1) \, \mathrm{d}x.$

```python
# Mencari integral tanpa batas dari fungsi

x = sy.Symbol('x')
f = 8*x**3 - x**2 + 5*x - 1

f.integrate(x)
```

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.patches import Polygon
%matplotlib inline

# Mendefinisikan fungsi g
def g(x):
    return 8*x**3 - x**2 + 5*x - 1

# Membuat array dari nilai x dengan rentang 0 hingga 11
x = range(0, 11)

# Mendapatkan nilai y untuk setiap nilai x
y = [g(a) for a in x]

fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(linestyle = ':')

# # Plot x terhadap g(x)
plt.plot(x,y, color='royalblue')

# Membuat daerah di bawah kurva yang ingin dihitung
ix = np.linspace(4, 8)
iy = g(ix)
verts = [(4,0)] + list(zip(ix, iy)) + [(8, 4)]
poly = Polygon(verts, facecolor='powderblue')
ax.add_patch(poly)

plt.show()
```

```python
i, e = integrate.quad(lambda x: 8*x**3 - x**2 + 5*x - 1, 4, 8)
print('%.2f' % i)
```

Luas daerah di bawah kurva $y = 8x^{3}-x^{2}+5x-1$ dari $x = 4$ hingga $x = 8$ adalah:

$$\int_4^8 (8x^{3}-x^{2}+5x-1) \, \mathrm{d}x = 2x^4 - \frac{x^3}{3} + \frac{5x^2}{2} - x\ \big|_4^8 = 7646.67$$
